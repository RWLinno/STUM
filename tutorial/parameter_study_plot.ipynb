```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the Excel file
file_path = 'param_study.xlsx'
xl = pd.read_excel(file_path, engine='openpyxl')

# Check the sheet names to ensure correct data is being read
sheet_names = xl.sheet_names
sheet_names
```

```python
# Load the specific sheet into a DataFrame
df = xl.parse('Sheet0')

# Display the DataFrame to understand its structure
# df.head(20)
```

```python
df1 = df.iloc[0:3, :].set_index(df.iloc[0:3, 0]).drop(df.columns[0], axis=1)
df2 = df.iloc[3:9, :].set_index(df.iloc[3:9, 0]).drop(df.columns[0], axis=1)
df3 = df.iloc[9:15, :].set_index(df.iloc[9:15, 0]).drop(df.columns[0], axis=1)
df1 = df1*100
df2 = df2*100
df3 = df3*100
print(df2)
```

```python
fig, axes = plt.subplots(1,3, figsize=(10, 4))

# Plotting the first subplot
df1.index = [0,1,2]
df1.iloc[:, 0].plot(ax=axes[0], marker='o')
df1.iloc[:, 1].plot(ax=axes[0], marker='^')
df1.iloc[:, 2].plot(ax=axes[0], marker='s')
axes[0].set_xticks(df1.index)
axes[0].set_xticklabels(["5%", "10%", "100%"])
axes[0].set_xlabel('Different Proportions of Dataset')
axes[0].set_ylabel('Performance Drop (%)')
axes[0].legend(loc='upper right')

df2.iloc[1:, 0].plot(ax=axes[1], marker='o')
df2.iloc[1:, 1].plot(ax=axes[1], marker='^')
df2.iloc[1:, 2].plot(ax=axes[1], marker='s')
axes[1].set_xticks(df2.index[1:])
axes[1].set_xlabel('Number of Nalls')
axes[1].set_ylabel('Performance Drop (%)')
axes[1].legend(loc='upper right')

df3.iloc[1:, 0].plot(ax=axes[2], marker='o')
df3.iloc[1:, 1].plot(ax=axes[2], marker='^')
df3.iloc[1:, 2].plot(ax=axes[2], marker='s')
axes[2].set_xticks(df3.index[1:])
axes[2].set_xlabel('Number of Embedding Dimensions')
axes[2].set_ylabel('Performance Drop (%)')
axes[2].legend(loc='upper right')

axes[0].grid(True)
axes[1].grid(True)
axes[2].grid(True)
plt.tight_layout()
# plt.show()

plt.savefig('case_study.pdf', dpi=300, bbox_inches='tight')
```
